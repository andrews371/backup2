```python
import glob # para mexer com vários arquivos varrendo todos com mesma extensão e transformando num grande data set
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt # mais usadas essa e a de baixo
import seaborn as sns
import warnings
import time

warnings.filterwarnings('ignore')
%matplotlib inline
matplotlib.style.use('default')
```

```python
# lista com o caminho + nome dos arquivos .csv (apenas caminho + nome.csv). não tem nenhuma informação extra

file_list = [f for f in glob.glob("*.csv")]
```

```python
# obs: o "encoding = 'unicode_escape'" é usado para não dá erro que daria em alguns casos qnd 
# se lê como dataframe e há algum problema de codificação, então isso evita o erro

df = pd.read_csv(file_list[21],encoding = 'unicode_escape') # df é um dataframe apenas para o arquivo arquivo 21 do file_list
display(df.shape) # mostra as linhas e colunas. Se usasse num for podia mostrar linhas e colunas de cada arquivo
df.drop(['Unnamed: 0'],inplace=True,axis=1)
df.head()
```

```python
# criando um dataframe para todos os arquivos do file_list e não apenas para um como feito anteriormente
# ou seja é uma lista de dataframes

df_list = [pd.read_csv(f,encoding = 'unicode_escape') for f in file_list]
```

```python
final_dataset = pd.concat(df_list, sort=False).reset_index(drop=True) # o reset_index é para remover a coluna de índice
final_dataset.shape
final_dataset.drop(['Unnamed: 0'], inplace=True, axis=1)
```

```python
# adicionando coluna golscasa e golsvisitante

for key,row in final_dataset.iterrows():
    final_dataset.loc[key,'golscasa'] = row['placar'].split('x')[0]
    final_dataset.loc[key,'golsvisitante'] = row['placar'].split('x')[1]
display(final_dataset.head())
```

```python
# adicionando um id para cada jogo do mesmo dia

data_equipes = []
cont = 0
cont2 = 0
identificador = ''

for key, row in final_dataset.iterrows():
    if (row['data'] not in data_equipes) or (row['times'] not in data_equipes):
        data_equipes.append(row['data'])
        data_equipes.append(row['times'])
        cont += 1
        identificador = str(cont)
        data_equipes.append(identificador)
        final_dataset.loc[key, 'id'] = identificador
        cont2 += 1
    else:
        data = row['data']
        times = row['times']
        for i in range(len(data_equipes)):
            time.sleep(0.1)
            if (( data == data_equipes[i]) and ( times == data_equipes[i+1])):
                cont2 += 1
                final_dataset.loc[key, 'id'] = data_equipes[i+2]
                break
```
