```python
for key,row in team_df.iterrows():
    possebolacasaft = row['possebolacasaft']
    possebolavisitanteft = row['possebolavisitanteft']
    possebolacasaht = row['possebolacasaht']
    possebolavisitanteht = row['possebolavisitanteht']
    
    if possebolacasaft[len(possebolacasaft) - 1] == '%':
        possebolacasaft = possebolacasaft[:-1]
        team_df.loc[key,'possebolacasaft'] = possebolacasaft
    if possebolavisitanteft[len(possebolavisitanteft) - 1] == '%':
        possebolavisitanteft = possebolavisitanteft[:-1]
        team_df.loc[key,'possebolavisitanteft'] = possebolavisitanteft
    if possebolacasaht[len(possebolacasaht) - 1] == '%':
        possebolacasaht = possebolacasaht[:-1]
        team_df.loc[key,'possebolacasaht'] = possebolacasaht
    if possebolavisitanteht[len(possebolavisitanteht) - 1] == '%':
        possebolavisitanteht = possebolavisitanteht[:-1]
        team_df.loc[key,'possebolavisitanteht'] = possebolavisitanteht
```

```python
team_df = team_df.drop(['Unnamed: 0'], axis=1).sample(frac=1).reset_index(drop=True)
```

```python
team_df.to_csv('dados_durante_jogo_pre_processados.csv')
```
