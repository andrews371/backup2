```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression, ElasticNet
from sklearn.neighbors import KNeighborsRegressor
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import matplotlib
import seaborn as sns
matplotlib.style.use('ggplot')

seed = np.random.seed(42)
```

```python
from google.colab import drive
drive.mount('/content/drive')

df = pd.read_csv('/content/drive/My Drive/machine_learning_football_club/all_leagues_seasons.csv', sep=';', low_memory=False)
```

```python
# acima vemos que tem time com pontuaçao negativo por puniçao sofrida. mas para a gente nao interessa pontuaçao negativa
# entao vamos trabalhar so com times com pontuacoes positivas

df = df[df['points_for'] >= 0]
```

```python
# dropando strings que nao nos interessa
# apesar de ser mostrada uma mensagem estranha, a operaçao foi feita

df.drop(['team','league','season'], axis=1, inplace=True)
```

```python
df.sample(5)
```

```python
df.info()
```

```python
# aqui to usando seaborn com o matplotlib
# lembrando que variáveis que têm uma alta correlaçao significa que uma explica o resultado tanto quanto a outra
# entao podemos considerar apenas uma dessas variaveis
# podemos fazer isso atraves de uma maneira estatistica (melhor) como vai ser mostrado mais na frente

plt.figure(figsize=(10,10))
sns.set(font_scale=1.5)
sns.heatmap(df.corr(), cmap='RdBu_r')
plt.show()
```

```python
# variavel preditora
x = df.drop(['final_points'], axis=1) # aqui estou jogando todas as variaveis exceto "final_points" no x

# variavel alvo
y = df['final_points'].values.reshape(-1,1) # aqui estou jogando apenas a variavel "final_points" no y
```

```python
# escalando valores de entrada para ficar no mesmo range

scaler = StandardScaler()
scaled_x = scaler.fit_transform(x)
```

```python
# agora vou usar o método estatistico mencionado anteriormente, o pca
# onde as variaveis com o menor valor pouco influenciam no resultado ou nao influenciam
# e assim podem ser descartadas e, ao elimina-las diminui-se a complexidade do modelo e ele pode 
# trabalhar melhor, ter melhores respostas. entao a ideia é eliminar as que nao interferirem tanto no resultado

pca = PCA(15, random_state=seed) # esse 15 é pq temos 15 variaveis sem contar com a variavel alvo "final_points"
pca.fit(scaled_x)
variance = pca.explained_variance_ratio_
var = np.cumsum(np.round(pca.explained_variance_ratio_, decimals=7) * 100)
var
```

```python
# na lista acima vemos o seguinte: a 1a variavel explica 51.9% da variancia (interfere no resultado final 51.9%)
# a 2a e a 1a juntas explicam 82.15%, a 3a, 2a e 1a juntas explicam 89.4% e assim sucetivamente (é uma soma acumulado da atual com a anterior)
# entao até a 7a variavel temos 99.09% do resultado final explicado. entao vamos focar apenas nessas 7
# para diminuir a complexidade do modelo como ja falado anteriormente
```

```python
# plotando o numero de componentes (variaveis) que definimos (15)

fig, ax1 = plt.subplots(ncols=1, figsize=(10,8))
ax1.set(xlabel='Número de Componentes', ylabel='Variância Explicada (%)', title='Análise do PCA', ylim=[68,101], xticks=list(range(1,16)))
sns.lineplot(x=list(range(1,16)), y=var, ax=ax1)
plt.show()
```

```python
# no grafico acima vemos que a medida que tenho mais variaveis mais a explicaçao do resultado final se aproxima de 100%
# e como percebemos na lista gerada pelo PCA anteriormente, com 7 variaveis ja temos um alto indice de explicaçao (99.09%)

# enquanto antes definimos o numeros de variaveis que queriamos (15) agora vamos definir o percentual ate que percentual de explicaçao
# queremos a variavel

# O ".99" é 99%
pca = PCA(.99, random_state=seed)
pca.fit(scaled_x)
data = pca.transform(scaled_x)
print(f'Percentual de variancia explicada por cada componente: {np.round(pca.explained_variance_ratio_*100,3)}')
print(f'Perentual total de variancia explicada pelas componentes: {np.round(pca.explained_variance_ratio_*100,3).sum():.3f}%')
```

```python
# divisao em treino e testes
x_train, x_test, y_train, y_test = train_test_split(data, y, test_size=0.2)
```

## LinearRegression()

```python
model = LinearRegression()

# com o metodo fit() fazemos o treinamento usando os dados de treino de entrada (x_train) e dados de treino de saida (y_train)
model.fit(x_train, y_train)

# para visualizar os coeficientes encontrados
print(f'Coeficiente {model.coef_}')
print(f'Constante {model.intercept_}')

# com o metodo predict() fazemos a previsão
y_pred = model.predict(x_test)

# exibe o R² 
# o objetivo é maximizar o R², pois enquanto mais proximo de 1, melhor é o modelo para explicar
# o resultado
print(f'1a maneira de visualizar o R2 {model.score(x_test, y_test):.4f}')
print(f'2a maneira de visualizar o R2 {r2_score(y_test, y_pred):.4f}')

# calcula e exibe o erro medio quadrático
# enquanto mais proximo de 0, melhor
print(f'Erro medio quadrado {mean_squared_error(y_test, y_pred):.4f}')

# calcula e exibe a raiz do erro medio quadrático
print(f'Raiz do erro medio quadrado {np.sqrt(mean_squared_error(y_test, y_pred)):.4f}')
```

```python
final_data = pd.DataFrame(np.hstack((y_test, y_pred)))
final_data.rename(index=str, columns={0:'Actual', 1:'Prediction'}, inplace=True)
fig, ax1 = plt.subplots(ncols=1, figsize=(10,8))
sns.scatterplot(x='Actual', y='Prediction', data=final_data, ax=ax1, color='green')
ax1.set(xlabel='Pontuação Final Verdadeira', ylabel='Pontuação Final predita pelo modelo', title='Pontuação Final - Verdadeira x Predita')
plt.show()
```

## RandomizedSearchCV()

```python
# obs.: parametros sao valores intrinsecos do modelo que nao temos controle sobre ele. Exemplo coeficientes em regressao linear
# pesos em quando estamos tratando com rede neural. Coisas que o modelo retorna e calcula por si so a partir dos dados mas nao temos o controle sobre isto
# hiperparametros sao os parametros que podemos ajustar no modelo. Exemplo em uma rede neural uma caixa de aprendizado, no KNN o numero 
# de K vizinhos. Sao valores que ajustamos para melhorar o modelo

from sklearn.model_selection import RandomizedSearchCV
params = {'alpha':[0.5, 1, 1.5, 2, 2.5], 'l1_ratio':[0.1, 0.2, 0.5, 0.8, 1], 'warm_start':[True, False]}

en = ElasticNet(selection = 'random', precompute = True, max_iter = 10000)
model = RandomizedSearchCV(en, param_distributions = params, cv = 10, random_state = seed)

model.fit(x_train, y_train)
model.best_params_
```

## Validação Cruzada

```python
from sklearn.model_selection import KFold, cross_val_score

# prepare models
models = []

# o ideal é usar os melhores parametros que encontramos para cada algoritmo quando fizemos a hiperparametrizaçao
# aqui nao usei no caso do KNN, mas foi para seguir o video (nao sei se ele nao usou porque se confundiu ou foi para ganhar tempo)
models.append(('LRM', LinearRegression()))
models.append(('ENM', ElasticNet(warm_start=False, l1_ratio=0.2, alpha=0.5, max_iter=10000, selection = 'random', precompute = True)))
models.append(('KNN', KNeighborsRegressor(n_neighbors=35, weights='uniform', metric='euclidean', algorithm='brute')))

# evaluate each model in turn
results = []
names = []
for name, model in models:
  kfold = KFold(n_splits=10, random_state=seed)
  cv_results = cross_val_score(model, data, y, cv=kfold)
  results.append(cv_results)
  names.append(name)
  print(f'{name}: R2 medio {cv_results.mean():.4f}, Desvio padrão do R2 {cv_results.std():.4f}')

# Boxplot para comparar os modelos
fig, ax1 = plt.subplots(ncols=1, figsize=(10, 8))
sns.boxplot(x=names, y=results, ax=ax1)
ax1.set(xlabel='Modelo', ylabel='Coeficiente R2', title='Comparação dos modelos')
plt.show()
```

## Previsao campeonato brasileiro 2019

```python
def get_table(current_round,max_rounds,league,season,flag):
  dfs = pd.read_html(f'https://www.worldfootball.net/schedule/{league}-{season}-{flag}{current_round}', header=0)
  df = dfs[3]
  # agora vamos apagar a coluna "Team" que aqui não interessa
  df.drop(['Team'], inplace=True, axis=1)

  df.rename(index=str, columns={'#':'pos', 'Team.1':'team', 'M.':'current_round', 'W':'win',
                                'D':'draw', 'L':'loss', 'goals':'goals', 'Dif.':'goals_diff', 'Pt.':'points_for'}, inplace=True)
 
  aux = df['goals'].str.split(':', n=1, expand=True)
  df['goals_for'] = aux[0]
  df['goals_against'] = aux[1]
  df.drop(columns = ['goals'], inplace=True)
  df = df.apply(pd.to_numeric, errors='ignore')
  df['pos'] = pd.to_numeric(df.index) + 1
  df['max_rounds'] = max_rounds
  df['rounds_left'] = df['max_rounds'] - df['current_round']
  df['possible_points'] = df['current_round'] * 3
  df['performance'] = np.round((df['points_for'] / df['possible_points']) * 100, 2)
  df['performance'].replace(np.inf, df['points_for'], inplace=True)
  df.fillna(value={'performance':0}, inplace=True)
  df['total_possible_points'] = df['points_for'] + (df['rounds_left'] * 3)
  df['goals_for_against_ratio'] = df['goals_for'] / df['goals_against']
  df['goals_for_against_ratio'].replace(np.inf, df['goals_for'], inplace=True)
  df.fillna(value={'goals_for_against_ratio':0}, inplace=True)
  df['goals_for_against_ratio'] = np.round(df['goals_for_against_ratio'], 3)
  df['league'] = league
  df['season'] = season
  columns = ['pos', 'team', 'current_round', 'max_rounds', 'rounds_left', 'win', 'loss', 'draw',
             'goals_for', 'goals_against', 'goals_diff', 'goals_for_against_ratio', 'points_for', 'possible_points',
            'total_possible_points', 'performance', 'league', 'season']
  df = df[columns]
  return df
```
