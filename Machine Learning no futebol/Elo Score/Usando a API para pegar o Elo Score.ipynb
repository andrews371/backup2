## Usar a API para pegar o Elo Score

```python
from io import StringIO
list_df = [] # lista que vai receber dataframes

# os dados recebidos vão ser colocados em um dataframe (elo_df) e o dataframe vai para uma lista de dataframes
# essa lista é a list_df, uma lista de dataframes
for data in team_df.Date.unique():
    elo_csv = requests.get(f"http://api.clubelo.com/{data}", allow_redirects=True)
    elo_csv = str(elo_csv.content,'utf-8')
    elo_csv = StringIO(elo_csv)
    elo_df = pd.read_csv(elo_csv)
    list_df.append(elo_df)
```

```python
elo_df_final = pd.concat(list_df)
```

## Criando colunas com Elo Score do mandante e do visitante

```python
for key,row in copy_team_df.iterrows():
    
    # substitui o nome "Ath Madrid" do meu df "copy_team_df" por Atletico para ficar como está 
    # no df "elo_df_esp"
    if row['HomeTeam'] == 'Ath Madrid':
        copy_team_df.loc[key,'HomeTeam'] = 'Atletico'
        
    # elimina "La Coruna" do df "copy_team_df" para ficar como está no df "elo_df_esp"
    # lá não tem o La Coruna nem com outro nome
    if row['HomeTeam'] == 'La Coruna':
        copy_team_df.drop(key, inplace=True) # eliminando o La Coruna de copy_team_df
        
    choices = list(elo_df_esp.Club.unique())
    team = process.extract(row['HomeTeam'], choices, limit=2)[0][0] # o célula anterior exemplifica esse "[0][0]"
    team_elo = elo_df_esp[(elo_df_esp['Club'] == team) & ((elo_df_esp['To'] == row['Date']) | (elo_df_esp['From'] == row['Date']))]
    if not team_elo.empty:
        copy_team_df.loc[key,'home_elo'] = team_elo.values[0][4]
```

```python
for key,row in copy_team_df.iterrows():
    
    # substitui o nome "Ath Madrid" do meu df "copy_team_df" por Atletico para ficar como está 
    # no df "elo_df_esp"
    if row['AwayTeam'] == 'Ath Madrid':
        copy_team_df.loc[key,'AwayTeam'] = 'Atletico'
        
    # elimina "La Coruna" do df "copy_team_df" para ficar como está no df "elo_df_esp"
    # lá não tem o La Coruna nem com outro nome
    if row['AwayTeam'] == 'La Coruna':
        copy_team_df.drop(key, inplace=True) # eliminando o La Coruna de copy_team_df
        
    choices = list(elo_df_esp.Club.unique())
    team = process.extract(row['AwayTeam'], choices, limit=2)[0][0] # o célula anterior exemplifica esse "[0][0]"
    team_elo = elo_df_esp[(elo_df_esp['Club'] == team) & ((elo_df_esp['To'] == row['Date']) | (elo_df_esp['From'] == row['Date']))]
    if not team_elo.empty:
        copy_team_df.loc[key,'away_elo'] = team_elo.values[0][4]
```
