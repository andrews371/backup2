```python
# Entendendo a função Apply

import pandas as pd
import numpy as np

# configurando o arrendondamento em 2 casas decimais
pd.options.display.float_format = '{:,.2f}'.format

valor_apostado = np.array([10, 30, 20, 20, 10, 20]) 
odds_apostadas = np.array([1.8, 2.20, 2.05, 2.75, 1.75, 2.50])
lista_de_apostas = np.array(['Atlético - MG', 'Cruzeiro', 'Palmeiras - 0.75', 'Mais de 2,5 Gols', 'Cartões Amarelos + 5', 'Cantos + 12,5'])

apostas_df = pd.DataFrame({'Valor apostado':valor_apostado, 'Odds apostadas':odds_apostadas, 'Apostas':lista_de_apostas})

apostas_df['lucro_potencial'] = (apostas_df['Odds apostadas'] - 1) * apostas_df['Valor apostado']

apostas_df['lucro_potencial'] = (apostas_df['Odds apostadas'] - 1) * apostas_df['Valor apostado']

apostas_df['prob_aprox'] = 1 / apostas_df['Odds apostadas']

apostas_df = apostas_df[['Apostas', 'Odds apostadas', 'Valor apostado', 'lucro_potencial', 'prob_aprox']]

display(apostas_df) # podia nao usar o display mas é melhor usar pq sem usar pode n funcionar se tiver muitos dados
```

```python
# para dobrar o valor apostado eu poderia fazer:

apostas_df['aposta_dobrada'] = apostas_df['Valor apostado'] * 2
display(apostas_df)
```

```python
# mas posso usar o "apply"

def dobrar_aposta(x):
    '''multiplica o valor de x por 2'''
    return x * 2
```

```python
del apostas_df['aposta_dobrada']
display(apostas_df)
```

```python
# agora que temos nossa função podemos fazer como segue
# lembrando que pra esse caso específico seria mais simples da primeira maneira
# mas pra outros casos o método apply é mais indicado e facilita bastante

# dobrando a aposta através do método apply


apostas_df['aposta_dobrada'] = apostas_df['Valor apostado'].apply(dobrar_aposta)
display(apostas_df)
```

```python
# assim ele gera o mesmo número aleatório para a coluna num_ale
apostas_df['num_ale'] = np.random.uniform()
display(apostas_df)
```

```python
# assim os números aleatórios são diferentes
# vamos usar funções lambda ou funções anônimas

apostas_df['num_ale'] = ''
apostas_df['num_ale'] = apostas_df['num_ale'].apply(lambda x: np.random.uniform())
display(apostas_df)
```

```python
# agora vamos usar o apply não apenas em linhas de uma coluna mas nas linhas das colunas informadas
# se faz assim quando queremos usar mais de uma coluna (variável) nas operações

apostas_df['resultado'] = apostas_df.apply(lambda row: 'ganha' if row['num_ale'] <= row['prob_aprox'] else 'perdida', axis = 1)
display(apostas_df)
```

```python
# verificando o lucro e o prejuízo

def gerar_resultado(val_apostado, luc_potencial, resultado):
    '''retorna o valor após a aposta quanto de lucro ou prejuízo'''
    if resultado == 'ganha':
        lucro_preju = luc_potencial
    else:
        lucro_preju = val_apostado * (-1)
    return lucro_preju
```

```python
apostas_df['lucro_prejuizo'] = apostas_df.apply(lambda row: gerar_resultado(row['Valor apostado'], row['lucro_potencial'], row['resultado']), axis = 1)
display(apostas_df)
```
