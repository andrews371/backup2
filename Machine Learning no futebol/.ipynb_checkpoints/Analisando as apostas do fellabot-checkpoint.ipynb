```python
# verificando o lucro e o prejuízo

def gerar_resultado(val_apostado, odds, resultado):
    '''retorna o valor após a aposta quanto de lucro ou prejuízo'''
    if resultado == 'Sim':
        lucro_preju = val_apostado * (odds - 1)
    else:
        lucro_preju = val_apostado * (-1)
    return lucro_preju
```

```python
fellabot['Resultado'] = fellabot.apply(lambda row: gerar_resultado(row['Valor da Entrada'], row['Odd'], row['Ganhou a aposta?']), axis = 1)
```

## Utilizando o pandas iterrows
## Obtendo o mandante e visitante

```python
fellabot.head()
```

```python
# quebra a string anterior por "/" pega o último valor que é o nome dos times
# quebra por "-x-" que é o que separa os times 
fellabot.loc[346, 'Partida'].split('/')[-1].split('-x-')
```

```python
# agora vamos remover esse "-" entre o nome dos times e vamos colocar a primeira letra em maiúsculo

# posso fazer para o outro time tbm. usei o capitalize. aqui só a 1a letra da 1a palavra fica com letra maiúscula.
fellabot.loc[346, 'Partida'].split('/')[-1].split('-x-')[-1].replace('-', ' ').capitalize()
```

```python
# aqui uso o title. todas as 1as letras de cada palavra ficam maiúsculas

fellabot.loc[346, 'Partida'].split('/')[-1].split('-x-')[-1].replace('-', ' ').title()
```

```python
# agora vamos obter mandante e visitante para toda a tabela (usando o row)
# e também já criamos duas novas colunas: "Mandante" e "Visitante" (usando o key)

for key, row in fellabot.iterrows():
    fellabot.loc[key, 'Mandante'] = row['Partida'].split('/')[-1].split('-x-')[0].replace('-', ' ').capitalize()
    fellabot.loc[key, 'Visitante'] = row['Partida'].split('/')[-1].split('-x-')[1].replace('-', ' ').capitalize()
```

## mandante e visitante com apply ( mais fácil )

```python
# aqui vamos apagar a coluna "Mandante" e "Visitante" criada anteriormente para refazer utilizando apply
# aproveitamos para aprender a fazer isso com a função drop

fellabot.drop(['Mandante', 'Visitante'], axis=1) # assim ele apenas exibe sem essas colunas mas não tirou de fato
```

```python
# verificando que não tirou

fellabot.head()
```

```python
# agora usando o drop e eliminando de fato as colunas "Mandante" e "Visitante"
# para isso usamos o inplace como argumento além do axis que já usávamos

fellabot.drop(['Mandante', 'Visitante'], inplace=True ,axis=1)
```

```python
fellabot.head()
```

## agora vamos criar novamente aquelas colunas usando o apply

```python
def casa_fora(url,time):
    '''pegando da url o time da casa e o visitante dos times'''
    
    if time == 'mandante':
        nome = url.split('/')[-1].split('-x-')[0].replace('-', ' ').capitalize()
        
    else:
        nome = url.split('/')[-1].split('-x-')[1].replace('-', ' ').capitalize()
        
    return nome
```

```python
fellabot['Mandante'] = fellabot.apply(lambda row: casa_fora(row['Partida'], 'mandante'), axis=1)
fellabot['Visitante'] = fellabot.apply(lambda row: casa_fora(row['Partida'], 'visitante'), axis=1)
```
