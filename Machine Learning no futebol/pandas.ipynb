```python
apostas_df['prob_aprox'] = 1 / apostas_df['Odds apostadas']
display(apostas_df)
```

```python
apostas_df['prob_aprox'] = np.around(1 / apostas_df['Odds apostadas'], 2)
display(apostas_df)
```

```python
# para não ter que usar esse "np.around()" toda vez que quisermos arredondar pra 2 casas decimais
# configuramos o pandas como este comando "pandas.options.display.float_format = '{:,.2f}'.format"
# antes

apostas_df['prob_aprox'] = 1 / apostas_df['Odds apostadas']
display(apostas_df)
```

```python
# após uso do comando de configuração mencionado acima...

# este comando só precisa ser digitado uma vez no começo do código, coloquei aqui
# apenas como exemplificação mas o ideal é no começo do código

pd.options.display.float_format = '{:,.2f}'.format

apostas_df['prob_aprox'] = 1 / apostas_df['Odds apostadas']
display(apostas_df)
```
